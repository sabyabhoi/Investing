```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_ta as ta

import os

import utils
```

```python
DATA_PATH = '../../data'
```

```python
tickers = pd.read_csv(os.path.join(DATA_PATH, 'tickers/ind_niftymidcap100list.csv'))
tickers = (tickers['Symbol']+'.NS').to_list()
tickers
```

```python
df = utils.download_and_cache(tickers, os.path.join(DATA_PATH, 'stocks'))
df
```

```python
df_close = pd.concat([df[t]['Adj Close'].rename(t) for t in tickers], axis=1).dropna()
df_close.tail()
```

```python
df_rsi = pd.concat([df_close[t]
 .rename('close')
 .to_frame()
 .ta.rsi(length=7)
 .dropna()
 .rename(t) for t in tickers], axis=1)
df_rsi.tail()
```

```python
current_tickers = df_rsi.apply(lambda row: row.nlargest(20).index.to_list(), axis=1).iloc[-1]
current_tickers
```

```python
df1_close = df_close[current_tickers]
df1_close
```

### Actual Optimization

```python
from skfolio import Population, RiskMeasure, Portfolio
from skfolio.optimization import ObjectiveFunction, MeanRisk, EqualWeighted
```

```python
def bootstrap(iter = 100, selection_percentage=0.6):
    model = MeanRisk(risk_measure=RiskMeasure.VARIANCE,
                    objective_function=ObjectiveFunction.MAXIMIZE_RATIO,
                    portfolio_params=dict(name='Max Sharpe'))
    w = []
    for _ in range(iter):
        indices = np.random.choice(df1_close.shape[0], 
                                int(df1_close.shape[0] * selection_percentage), 
                                replace=True)
        portfolio = model.fit_predict(df1_close.reset_index(drop=True).loc[indices,current_tickers]
                                    .pct_change()
                                    .dropna())
        w.append(portfolio.weights)
    w = np.array(w).T
    w = w @ np.ones(shape=(iter, 1))
    return (w / np.sum(w)).flatten()
bootstrap_weights = bootstrap(500, 1.0)
bootstrap_weights
```

```python
pop = Population([Portfolio(
    df1_close.pct_change().dropna(), 
    bootstrap_weights,
    name='Sharpe Maximization'
    ),
    EqualWeighted().fit_predict(df1_close.pct_change().dropna())])
pop.summary()
```
